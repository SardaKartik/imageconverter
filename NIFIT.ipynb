```python
def image_nifty_to_png(nifty_path, image_path, study_name, series_name):
    img = nib.load(nifty_path)
    data = img.get_fdata()
    
    os.makedirs(image_path, exist_ok=True)
    
    for i in range(data.shape[2]):
        plt.imshow(data[:, :, i], cmap="gray")
        plt.axis('off')
        file_name = f"{study_name}_{series_name}_s{i+1}.png"
        plt.savefig(os.path.join(image_path, file_name), bbox_inches='tight', pad_inches=0)
        plt.close()
```
